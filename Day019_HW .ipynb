```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://airtw.epa.gov.tw/CHT/Default.aspx")

selectSite = Select(browser.find_element_by_id("ddl_county"))
selectSite.select_by_value('Kaohsiung')
selectYear = Select(browser.find_element_by_id("ddl_site"))
selectYear.select_by_value('10')


# browser.find_element_by_id('ctl05_btnQuery').click()
```

```python
# 取得資料（等瀏覽器資料出現後才執行）
html_source = browser.page_source
html_source
```

```python
soup = BeautifulSoup(html_source, 'html.parser')
b = soup.select('.allif')[0].select('li>b')
ul = soup.select('.allif')[0].select('li')
data={}
data2=[]
for j in range(len(b)):
    if(j%3==0 or j==14):
        if(j!=15):
            data[b[j].text]=[]
            data2.append([])
        else:
            data2[len(data2)-1].append(b[j].text)
    else:
        data2[len(data2)-1].append(b[j].text)
data2=list(reversed(data2))
for key,item in data.items():
    for i in data2:
        data[key]=i
    data2.pop()
    
for i in ul:
    num = i.select('li>span')
    for i in num:
        print(i)
data
    
```
