### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
soup = BeautifulSoup(r.text,'html5lib')
# soup.find(class_='part_list_2').find_all('h3')
h3 = soup.select('.part_list_2')[0].select('h3')
for i in h3:
    print(i.select('.date')[0].text,end=' ')
    print(i.select('.tag')[0].text,end=' ')
    print(i.select('a')[0].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
d = {}
for i in h3:
    if i.select('.tag')[0].text not in d.keys():
        d[i.select('.tag')[0].text] = []
    title = i.select('a')[0].text
    d[i.select('.tag')[0].text].append(title)
d
```

### ③ 哪一個類別的文章最多

```python
maxs = 0
maxText = '';
for k,v in d.items():
    if len(v)>maxs:
        maxs = len(v)
        maxText = k
print(maxText)
d[maxText]
```
