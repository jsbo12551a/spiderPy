## 作業目標

1. 取出今天所有的新聞

## 1. 取出今天所有的新聞

```python

from selenium import webdriver
from bs4 import BeautifulSoup
import time

browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")
    

    
```

```python

from datetime import datetime, timedelta
yesterday = datetime.today().replace(hour=0, minute=0, second=0, microsecond=0)
terday_news=[]
print(yesterday)
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
h3 = soup.select('.part_list_2')[0].select('h3')
#     browser.execute_script("window.scrollTo(0, 10000);")
for i in h3:
    time = i.find(class_="date").text
    title = i.find('a').text
    news_time = datetime.strptime(time, '%Y/%m/%d %H:%M')
#     print(news_time,title)
    if(news_time<yesterday):
        break
    terday_news.append({'date':time,'news':title})
print(terday_news)   
```

## 2. 取出現在時間兩小時內的新聞

```python
tow = datetime.now() - timedelta(hours=2)
toe_news=[]
for i in h3:
    time = i.find(class_="date").text
    title = i.find('a').text
    news_time = datetime.strptime(time, '%Y/%m/%d %H:%M')
#     print(news_time,title)
    if(news_time<tow):
        break
    toe_news.append({'date':time,'news':title})
print(toe_news)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
three_d_ago = datetime.now() - timedelta(days=3)
range_start = three_d_ago.replace(hour=15, minute=0, second=0, microsecond=0)
range_end = three_d_ago.replace(hour=17, minute=0, second=0, microsecond=0)

three_news=[]
for i in h3:
    time = i.find(class_="date").text
    title = i.find('a').text
    news_time = datetime.strptime(time, '%Y/%m/%d %H:%M')
#     print(news_time,title)
    if(range_start>news_time<range_end):
        three_news.append({'date':time,'news':title})
print(three_news)
```
